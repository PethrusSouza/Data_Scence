```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            "McDonald's", 'Visa',
            'Facebook', 'Alibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273  ]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
```

```python
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhes")
```

```python
#redimencionando
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhes")
plt.show()
```

```python
#conhecendo o frontsize
plt.plot(empresas,bilhoes)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel("Bilhes", fontsize=20)
plt.show()
```

```python
#mudando a cor e legendando
plt.plot(empresas,bilhoes, color = "green")
plt.legend(["Bilhes de dlares"], loc = "upper right", fontsize=15)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel("Bilhes", fontsize=20)
plt.show()
```

```python
#aumentar a espessura da linha (lw)
plt.plot(empresas,bilhoes, color = "green", lw=5)#lw = largura da linha
plt.legend(["Bilhes de dlares"], loc = "upper right", fontsize=15)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel("Bilhes", fontsize=20)
plt.show()
```
